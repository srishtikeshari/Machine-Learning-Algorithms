# Decision Trees in Python

### Import Libraries

```python
import pandas as pd
```

### Get The Data. Load data "spambase_data.csv"

```python
df = pd.read_csv('spambase_data.csv',header = None)
df.head() # used to display top 5 rows
```

### Train and Test Split

```python
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
```

```python
X = df[df.columns[:-1]]
Y = df[len(df.columns)-1]
```

#### Obtain X_train, X_test, Y_train, Y_test by splitting the dataset into 70-30 ratio with a random state value 30.

Note: random_state is used to control the shuffling in the split of the dataset.

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30, random_state = 30)
```

### Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(X_train,Y_train)
```

### Prediction and Evaluation

```python
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
predictions = dtree.predict(X_test)
```

```python
print(classification_report(Y_test,predictions))
print("Confusion Matrix")
print(confusion_matrix(Y_test,predictions))
print("\n Accuracy")
accuracy_gini = accuracy_score(Y_test, predictions)
print(accuracy_gini)
```

### Tree Visualization 

```python
import matplotlib.pyplot as plt
from sklearn import tree
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (16,16), dpi=1000)
tree.plot_tree(dtree, feature_names = df.columns)
fig.savefig('decision_tree.png')
```

#### Q1: Create a decision stump and find the majority class in the dataset

```python
#### Code here ####
import matplotlib.pyplot as plt
from sklearn import tree
stump = DecisionTreeClassifier(max_depth = 1, random_state=30)
stump.fit(X_train,Y_train)
pred_stump = stump.predict(X_test)
print(classification_report(Y_test,pred_stump))
print("Confusion Matrix")
print(confusion_matrix(Y_test,pred_stump))
print("\n Accuracy")
accuracy_stump = accuracy_score(Y_test,pred_stump)
print(accuracy_stump)
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (16,16), dpi=1000)
tree.plot_tree(stump, feature_names = df.columns)
fig.savefig('decision_tree_stump.png')
print(pred_stump)

```

```python
#### Majority class ####
class1 = (pred_stump > 0).sum()
class0 = (pred_stump < 1).sum()
if class1 > class0:
    print("Majority is Class 1")
else:
    print("Majority is Class 0")

```

#### Q2: Implement the decision tree classifier using entropy to measure the quality of split and random state value as 30. 
Note: The features are always randomly permuted at each split, the best found split may vary across different runs. To obtain a deterministic behaviour during fitting, random state has to be fixed to an integer.

```python
#### Code here ####
dtree_entropy = DecisionTreeClassifier(criterion='entropy',random_state=30)
dtree_entropy.fit(X_train,Y_train)
predictions_entropy = dtree_entropy.predict(X_test)
#fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (16,16), dpi=1000)
#tree.plot_tree(dtree_entropy, feature_names = df.columns)
#fig.savefig('decision_tree_entropy.png')
print(classification_report(Y_test,predictions_entropy))
print("Confusion Matrix")
print(confusion_matrix(Y_test,predictions_entropy))
print("\n Accuracy")
accuracy_entropy = accuracy_score(Y_test, predictions_entropy)
print(accuracy_entropy)
```

#### Q3: Compare the two trees on the basis of their accuracy. Which tree gives better accuarcy (using gini index or entropy).

```python
#### Write your answer here ####
if accuracy_gini > accuracy_entropy:
    print("Decision tree with Gini is better")
else:
    print("Decision tree with Entropy is better")
```

#### Q5: Find the feature importance from the trained model and rebuild the model using top 10 important features only. 

```python
#### Code here ####
importance = dtree_entropy.feature_importances_
imp_array = []
for i,v in enumerate(importance):
    imp_array.append((v,i))
imp_array.sort(reverse=True) 
col_top10 = []
for i in range(0,11):
    col_top10.append(imp_array[i][1])
print(col_top10)
```

#### Q6: Is there any information loss in the tree created above

```python
#### Write your answer here ####
X_top10 = df[df.columns[col_top10]]
Y_top10 = df[len(df.columns) - 1]
X_train_top10, X_test_top10, Y_train_top10, Y_test_top10 = train_test_split(X_top10,Y_top10,test_size=0.30, random_state=30)
dtree_top10 = DecisionTreeClassifier(criterion='entropy', random_state=30)
dtree_top10.fit(X_train_top10, Y_train_top10)
predictions_top10 = dtree_top10.predict(X_test_top10)
print(classification_report(Y_test_top10,predictions_top10))
print("\nConfusion Matrix")
print(confusion_matrix(Y_test_top10,predictions_top10))
print("\n Accuracy")
print(accuracy_score(Y_test_top10, predictions_top10))
print("\n\n")

if accuracy_score(Y_test, predictions_entropy) > accuracy_score(Y_test_top10, predictions_top10):
    print("Yes, there is information loss")
else:
    print("No information loss")
```
